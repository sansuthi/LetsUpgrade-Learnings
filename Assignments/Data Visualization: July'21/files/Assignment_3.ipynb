### 1. Plot using sns.relplot, line plot, box plot on fmri data set

```python
# Importing Libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
```

```python
# Loading 'fmri' dataset
fmri= sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.shape
```

## _RELPLOT_

```python
sns.relplot(x='timepoint', y='signal', data=fmri, hue='subject', style='event', size='region')
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='subject')
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()

sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='region')
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

## _LINEPLOT_

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='timepoint', y='signal', data=fmri, color='green', marker='*', markersize=15)
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
sns.lineplot(x='timepoint', y='signal', marker='o', data=fmri, hue='region')
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.subplot(1, 2, 2)
sns.lineplot(x='timepoint', y='signal', marker='o', data=fmri, hue='event')
plt.title('Timepoint vs Signal', fontsize=15)
plt.xlabel('TIME POINT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

## BOXPLOT

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
sns.boxplot(x='subject', y='signal', data=fmri)
plt.title('Signal: Subject-wise', fontsize=15)
plt.xlabel('SUBJECT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()

plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
sns.boxenplot(x='subject', y='signal', data=fmri)
plt.xlabel('SUBJECT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
sns.boxplot(x='subject', y='signal', data=fmri, hue='region')
plt.title('', fontsize=15)
plt.xlabel('SUBJECT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.subplot(1, 2, 2)

sns.boxenplot(x='subject', y='signal', data=fmri, hue='event')
plt.title('', fontsize=15)
plt.xlabel('SUBJECT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 1, 1)
sns.boxplot(x='region', y='signal', data=fmri, hue='subject')
plt.title('Signal: Region-wise', fontsize=15)
plt.xlabel('REGION', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.xticks(fontsize=15)
plt.subplot(2, 1, 2)
sns.boxenplot(x='event', y='signal', data=fmri, hue='subject')
plt.title('Signal: Event-wise', fontsize=15)
plt.xlabel('EVENT', fontsize=12)
plt.ylabel('SIGNAL', fontsize=12)
plt.xticks(fontsize=15)
plt.show()
```
