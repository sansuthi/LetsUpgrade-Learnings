### 1. plot a line plot between a and b:  
```
a = np.arange(40,50)  
b = np.arange(50,60)
```

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
a = np.arange(40, 50)  
b = np.arange(50, 60)
print('a:', a)
print('b:', b)
````

````python
plt.figure(figsize=(10,5))
plt.plot(a, b, 'g*--', markersize=10)
plt.title('LINE PLOT', color='b', fontsize=16)
plt.xlabel('a', color='r', fontsize=20)
plt.ylabel('b', color='r', fontsize=20)
plt.show()
````

### 2. Plot a line plot showing the sales trend in company 1 and 2:
```
days = [1,2,3,4,5,6,7] #days of the week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

````python
days = [1, 2, 3, 4, 5, 6, 7]
sales_1 = [160, 150, 140, 145, 175, 165, 180]
sales_2 = [70, 90, 160, 150, 140, 145, 175]
````

````python
plt.figure(figsize=(10,5))
plt.plot(days, sales_1, 'go--', markersize=7, label='company1')
plt.plot(days, sales_2, 'r*--', markersize=9, label='company2')
plt.title('SALES TREND', color='b', fontsize=16)
plt.xlabel('Week days', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.legend()
plt.show()
````

### 3.  Create a 3 by 3 subplots:
```
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

```

````python
x = [1, 2, 3, 4]
y1 = [4, 3, 2, 1]
y2 = [10, 20, 30, 40]
y3 = [40, 30, 20, 10]
y4 = [1, 2, 1, 2]
y5 = [40, 70, 90, 70]
````

````python
plt.figure(figsize=(12,9))
plt.subplot(3, 3, 1)
plt.plot(x ,y1,'b-')
plt.subplot(3, 3, 2)
plt.plot(x, y2,'g-.')
plt.subplot(3, 3, 3)
plt.plot(x, y3,'ro--')
plt.subplot(3, 3, 4)
plt.plot(x, y4,'b*-')
plt.subplot(3, 3, 5)
plt.plot(x, y5,'g*')
plt.subplot(3, 3, 6)
plt.plot(y1, y2,'rd-')
plt.subplot(3, 3, 7)
plt.plot(y2, y3,'bs')
plt.subplot(3, 3, 8)
plt.plot(y3, y4,'gh--')
plt.subplot(3, 3, 9)
plt.plot(y4, y5,'rD')
plt.show()
````
