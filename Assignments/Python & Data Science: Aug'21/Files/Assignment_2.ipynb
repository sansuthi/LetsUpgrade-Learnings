### 3. Do data analysis on two features of Boston data and write your insights.

```python
data = load_boston()
data
```

```python
boston = pd.DataFrame(data['data'])
boston.columns = data['feature_names']
boston['TARGET'] = data['target']
boston
```

```python
plt.figure(figsize=(25, 3))
plt.ylabel('TARGET')
for i, feat in enumerate(boston.columns[:-1]):
    plt.subplot(1, 13, i+1)
    plt.scatter(boston[feat], boston['TARGET'])
    plt.title(feat, fontsize=15)
plt.show()
```

```python
boston.corr()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(boston['LSTAT'], boston['TARGET'])
plt.title('LSTAT vs TARGET')
plt.xlabel('LSTAT')
plt.ylabel('TARGET')
plt.show()
```

* There is a linear relationship between variables

### 4. Apply simple linear regression

```python
# Predictor: LSTAT
X = boston[['LSTAT']]
y = boston['TARGET']
```

```python
# Scatterplot for LSTAT vs TARGET
plt.figure(figsize=(10,5))
plt.scatter(X, y)
plt.title('LSTAT vs TARGET', fontsize=18)
plt.xlabel('LSTAT', fontsize=15)
plt.ylabel('TARGET', fontsize=15)
plt.show()
```

```python
# Splitting Training & Testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
# Shape of each of train & test set
print(f'X_train: {X_train.shape}')
print(f'y_train: {y_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_test: {y_test.shape}')
```

```python
# Modeling & Fitting
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f'Coeffiecient: {model.coef_[0]}')
print(f'Intercept: {model.intercept_}')
```

```python
# Plotting Regression line on data
plt.figure(figsize=(10,5))
plt.title('Regression Line', fontsize=18)
plt.xlabel('LSTAT', fontsize=13)
plt.ylabel('TARGET', fontsize=13)
plt.scatter(X_train, y_train, color='orange', s=20, label='Train set')
plt.scatter(X_test, y_test, color='k', s=30, label='Test set')
plt.plot(X, model.coef_*X + model.intercept_, c='r', label='Regression Line')
plt.legend()
plt.show()
```

```python
# score of training the data
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Making Predictions on test data
y_pred = model.predict(X_test)
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
result.head(10)
```

```python
result.plot(kind='bar', figsize=(20, 5))
plt.title('Actual vs Predicted', fontsize=20)
plt.xlabel('Index', fontsize=15)
plt.ylabel('Target', fontsize=15)
plt.show()
```
